```python
#Exploratory Data Analysis

import pandas as pd
import matplotlib.pyplot as plt

#Load the "simulated" dataset
df = pd.read_csv("../data/simulated_gene_data.csv")
df.head()
```

```python
#GC% Histogram

plt.figure(figsize=(6,4))
df["gc_content"].plot(kind="hist", bins=20, edgecolor="black")
plt.xlabel("GC%")
plt.ylabel("Gene Count")
plt.title("Distribution of GC% Across Genes")
plt.tight_layout()
plt.savefig("../results/gc_hist.png", dpi=300)
plt.show()

```

```python
#Expression Histogram

plt.figure(figsize=(6,4))
df["expression"].plot(kind="hist", bins=20, edgecolor="black", color="orange")
plt.xlabel("Expression Level (a.u.)")
plt.ylabel("Gene Count")
plt.title("Distribution of Gene Expression Levels")
plt.tight_layout()
plt.savefig("../results/expression_hist.png", dpi=300)
plt.show()
```

```python
#Scatter Plot by Population

plt.figure(figsize=(6,5))
for pop in df["population"].unique():
    subset = df[df["population"] == pop]
    plt.scatter(subset["gc_content"], subset["expression"], s=30, alpha=0.7, label=pop)

plt.xlabel("GC%")
plt.ylabel("Expression Level (a.u.)")
plt.title("GC% vs Expression by Population")
plt.legend()
plt.tight_layout()
plt.savefig("../results/scatter_gc_vs_expression.png", dpi=300)
plt.show()
```
